## Concatenacin

La forma ms sencilla de combinar dos data sets es mediante concatenacin -al agregar un DataFrame al final de otro DataFrame. 

Pandas tiene la funcin <code>concat</code> para concatenar DataFrames.

Por defecto, pandas concatena filas en el eje horizontal


```python
sports_champions_D = pd.DataFrame(
                                data = [
                                        [2017, "Golden State Warriors"],
                                        [2018, "Golden State Warriors"],
                                        [2019,"Golden State Warriors"],
                                        [2020, "Los Angeles Lakers "]
                                        ],
                                columns = ["Year","Basketball"]
)
sports_champions_D
```

Podemos concatenar DataFrames de acuerdo al eje de las columnas, agregando el argumento <code>axis = 1</code> a la funcin <code>concat</code>

```python
sports_champions_F = pd.DataFrame(
                                    {"Hockey" : ["Pittsburgh Penguins", "Washington Capitals"]},
                                index = [2017, 2018]
)
sports_champions_F
```
